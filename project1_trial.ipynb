```python
# Applying melt (to a long format)
df=(df.melt(id_vars=["artiste","track","genre","date.entered","date.peaked"],var_name=["week"],value_name="rank"))
# Week to number 
#df.update(pd.DataFrame({"week": np.ravel([list(map(int, re.findall(r'\d+', i))) for i in df["week"]])}))
df
```

```python
df.update(pd.DataFrame({"week": np.ravel([list(map(int, re.findall(r'\d+', i))) for i in df["week"]])}))
df
```

```python
# creating a date column from date.entered and week
df['date.entered'] =  pd.to_datetime(df['date.entered'], format='%Y/%m/%d')
df=(df.assign(date= [df['date.entered'][i]+datetime.timedelta(weeks = df["week"][i]-1) for i in range(len(df["week"]))])
         .drop(['date.entered','date.peaked','week'], axis=1)
         .sort_values('artiste', ascending=True)
         .reset_index(drop=True))



df.head()
```

```python
# creating an id column from track
labels,unique=pd.factorize(list(df["track"]))
df["id"]=labels+1
df.reset_index(drop=True)
df.head()
```

```python
# creating a new dataframe for rank
df_rank=df.filter(["id", "date", "rank"]).dropna()
df_rank=df_rank.sort_values(by=['id','date']).reset_index(drop=True)

# creating a new dataframe for song
df_song=df.filter(["id", "artiste", "track"])
df_song=df_song.drop_duplicates('id').reset_index(drop=True)
```
