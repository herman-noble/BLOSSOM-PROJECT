#### modularisation

```python
car_ads=soup_vehicles.find_all('li', {'class': 'normal--2QYVk gtm-normal-ad'})
```

```python
car_ads
```

```python
Title=[]
Time_Posted = []
Mileage =[]
Location = []
Price=[]
```

```python
#loop over this page to get details of all other ads


for ad in car_ads:
    print(ad)
    print('\n\n\n')
```

```python
car_name=soup_vehicles.find('li', {'class': 'normal--2QYVk gtm-normal-ad'}).div.span.text

```

```python
for ad in car_ads:
    Title.append(ad.div.span.text)
    Mileage.append(ad.find('div', {'class':'content--3JNQz'}).div.div.text)
    Time_Posted.append(ad.find('div',{'class':'updated-time--1DbCk'}).text)
    Location.append(ad.find('div', {'class':'description--2-ez3'}).text.split(',')[0])
    Price.append(ad.find('div', {'class':'content--3JNQz'}).div.span.text)
```
